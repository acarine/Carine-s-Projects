```python
#Create data with only rooms, price, quadrant, assessment of neighboor, ward, square footage
df1 = df [['ROOMS','PRICE','CNDTN','QUADRANT','ASSESSMENT_NBHD','WARD','ZIPCODE','SQUARE']]
df1
```

```python
df1['QUADRANT'].dropna(inplace=True)
```

```python
#find any NaN
df1.isna().sum()
```

```python
df1.isnull().sum()
```

```python
#replace all missing price values with the mean
df1_filledPrices = df1.fillna(df1['PRICE'].mean())
```

```python
df1_filledPrices.head()
```

```python
df1_filledPrices['PRICE'].min()
```

```python
#Correlation between number of rooms and price
corr = df1_filledPrices.corr()
```

```python
print(corr)
```

```python
seaborn.heatmap(corr, vmin=-1, annot=True)
```

```python
#Grouping the houses by Quadrant --> Relationship of Price & Quadrant
df1_Price_Quadrant= df1_filledPrices.groupby('QUADRANT')['PRICE'].mean()
```

```python
df1_Price_Quadrant
```

```python
#Grouping houses by Condition ---> Relationship of Price & Condition
df1_filledPrices.groupby('CNDTN')['PRICE'].mean()
```

```python
plt.figure(figsize=(20,10))
seaborn.lmplot(x='ZIPCODE', y='PRICE', data=df1)
```

```python
#swarm plot of Stats from melted dataframe
seaborn.swarmplot(x='ZIPCODE', y='PRICE', 
              data=df1_filledPrices,
              hue='CNDTN')
```
